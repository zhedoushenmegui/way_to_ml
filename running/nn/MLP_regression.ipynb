```python
import numpy as np
import random
import pandas as pd
import math
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
```

```python
import os
project_path = os.path.abspath(os.path.join(os.path.dirname(os.path.abspath('.')), '../'))
import sys
sys.path.append(project_path)
from util.visualization import draw_lines
from util.visualization import draw_scatters
from util.evaluate_process import mse, rmse, mae
```

```python
pf = pd.read_csv('linear_model_data.params.10d.v1.csv')
df = pd.read_csv('linear_model_data.10d.v1.csv')
## 归一化
cf = df.copy()
mins = {k: min(df[k]) for k in df.columns}
maxs = {k: max(df[k]) for k in df.columns}
widths = {k: maxs[k] -mins[k] for k in df.columns}
###
for k in cf.columns:
    cf[k] = cf[k].map(lambda x: (x-mins[k])/widths[k])
cf.head()
```

```python
cf['d10'] = 1
feats = [f'd{i}' for i in range(11)]
Xtrain, Xtest, ytrain, ytest = train_test_split(cf[feats], cf.y)
len(Xtrain), len(Xtest)
```

```python
# 8.154444447490349e-05 0.009030196258936097 0.00363249103637447
mse_arr = [8.154444447490349e-05]
rmse_arr = [0.009030196258936097]
mae_arr = [0.00363249103637447]
times = [0]
```

```python
for i in range(1, 51):
    n_iter = i * 10 * 100
    clf = MLPRegressor(max_iter=n_iter, alpha=1)
    clf.fit(Xtrain, ytrain)

    pred = clf.predict(Xtest)
    rst_mse = mse(ytest, pred)
    
    rst_score = clf.score(Xtrain, ytrain)
    times.append(n_iter)
    mse_arr.append(rst_mse)
    mae_arr.append(mae(pred, ytest))
    rmse_arr.append(rmse(pred, ytest))
    #print(rst_mse, rmse(ytest, pred), mae(ytest, pred))
    #print(clf.score(Xtrain, ytrain), rst_score)
```

```python
draw_lines(times, [mse_arr, rmse_arr, mae_arr], ['mse', 'rmse', 'mae'])
```
